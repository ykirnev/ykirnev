*      _  _,       cmap_light  cmap_bold.      [](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

def plot_knn_bound(X, y, scaler=None, n_neighbors=10, xlim=(-15, 15), ylim=(-20, 20)):
    # step size in the mesh
    h = 0.05

    # Create color maps
    cmap_light = ListedColormap(['C0', 'orange', 'cyan', 'green'][:np.unique(y).shape[0]])
    cmap_bold = ['C0', 'orange', 'c', 'darkgreen'][:np.unique(y).shape[0]]

    x_min, x_max = xlim
    y_min, y_max = ylim
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    grid = np.c_[xx.ravel(), yy.ravel()]

    X_scaled = X # if scaler is None
    if scaler is not None:
        grid = scaler.transform(grid)
        X_scaled = scaler.transform(X)

    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, algorithm='brute')
    clf.fit(X_scaled, y)

    Z = clf.predict(grid)

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y,
                    palette=cmap_bold, alpha=1.0, edgecolor="black")
    plt.xlabel('x_1')
    plt.ylabel('x_2')
    plt.title('   {}-NN'.format(n_neighbors))
    plt.grid()
    plt.show()
```

   $1$-$NN$  $10$-$NN$

```python
plot_knn_bound(X, y, n_neighbors=1)
plot_knn_bound(X, y, n_neighbors=10)
```

```python
import pandas as pd
cnt_accuracies = []
tdif_accuracies = []

for params, score in results.items():
    if params[0] == 'CountVectorizer':
        cnt_accuracies.append(score)
    else:
        tdif_accuracies.append(score)
if cnt_accuracies:
    max_accuracy_cnt = max(cnt_accuracies)
    mean_accuracy_cnt = sum(cnt_accuracies) / len(cnt_accuracies)
else:
    max_accuracy_cnt = None
    mean_accuracy_cnt = None

if tdif_accuracies:
    max_accuracy_tdif = max(tdif_accuracies)
    mean_accuracy_tdif = sum(tdif_accuracies) / len(tdif_accuracies)
else:
    max_accuracy_tdif = None
    mean_accuracy_tdif = None

print("  CountVectorizer:")
print(f" : {max_accuracy_cnt}")
print(f" : {mean_accuracy_cnt}")

print("\n  TfidfVectorizer:")
print(f" : {max_accuracy_tdif}")
print(f" : {mean_accuracy_tdif}")

if max_accuracy_cnt is not None and max_accuracy_tdif is not None:
    if max_accuracy_cnt > max_accuracy_tdif:
        print("\nCountVectorizer    .")
    else:
        print("\nTfidfVectorizer    .")

    if mean_accuracy_cnt > mean_accuracy_tdif:
        print("CountVectorizer    .")
    else:
        print("TfidfVectorizer    .")
```

```python
accuracy_by_neighbors_cnt = {}
accuracy_by_neighbors_tdif = {}

for params, score in results.items():
    vectorizer_type = params[0]
    n_neighbors = params[1]
    if vectorizer_type == 'CountVectorizer':
        if n_neighbors not in accuracy_by_neighbors_cnt:
            accuracy_by_neighbors_cnt[n_neighbors] = []
        accuracy_by_neighbors_cnt[n_neighbors].append(score)
    elif vectorizer_type == 'TfidfVectorizer':
        if n_neighbors not in accuracy_by_neighbors_tdif:
            accuracy_by_neighbors_tdif[n_neighbors] = []
        accuracy_by_neighbors_tdif[n_neighbors].append(score)
average_accuracy_by_neighbors_cnt = {k: sum(v) / len(v) for k, v in accuracy_by_neighbors_cnt.items()}
average_accuracy_by_neighbors_tdif = {k: sum(v) / len(v) for k, v in accuracy_by_neighbors_tdif.items()}
neighbors_cnt = sorted(average_accuracy_by_neighbors_cnt.keys())
average_accuracy_cnt = [average_accuracy_by_neighbors_cnt[n] for n in neighbors_cnt]

neighbors_tdif = sorted(average_accuracy_by_neighbors_tdif.keys())
average_accuracy_tdif = [average_accuracy_by_neighbors_tdif[n] for n in neighbors_tdif]


plt.figure(figsize=(10, 6))
plt.plot(neighbors_cnt, average_accuracy_cnt, marker='o', label='CountVectorizer', color='blue')
plt.plot(neighbors_tdif, average_accuracy_tdif, marker='o', label='TfidfVectorizer', color='orange')
plt.title('      (K)')
plt.xlabel('  (n_neighbors)')
plt.ylabel('  (accuracy)')
plt.xticks(sorted(set(neighbors_cnt).union(neighbors_tdif)))
plt.grid()
plt.show()
```
