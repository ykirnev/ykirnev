````python
import time
import numpy as np

import research_functions_vectorised
import research_functions
````

````python
#  

are_multisets_equal_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(0, 2, n)
    y = np.random.randint(0, 2, n)
    are_multisets_equal_data += [(x, y)]

max_prod_mod_3_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 100, n)
    max_prod_mod_3_data += [x]

convert_image_data = []
for n in [10, 100, 1000]:
    image = np.random.random((n, n, 10))
    weights = np.random.random(10)
    convert_image_data += [(image, weights)]

rle_scalar_data = []
for n in [10, 100, 1000, 10000]:
    x = np.random.randint(1, 3, (n, 2))
    y = np.random.randint(1, 3, (n, 2))
    shapes = np.random.randint(1, 3, n)
    x[:, 1] = shapes
    y[:, 1] = shapes

    rle_scalar_data += [(x, y)]

cosine_distance_data = []
for n in [10, 100]:
    X = np.random.random((n, n))
    Y = np.random.random((n, n))
    cosine_distance_data += [(X, Y)]
````

````python
#     

def count_time(func, *data):
    tick = time.perf_counter()
    func(*data)
    return time.perf_counter() - tick
````

````python
#   
vectorized_times = []
non_vectorized_times = []

for data in are_multisets_equal_data:
    vec_time = count_time(research_functions_vectorised.are_multisets_equal, data[0], data[1])
    vectorized_times.append(vec_time)
    non_vec_time = count_time(research_functions.are_multisets_equal, data[0].tolist(), data[1].tolist())
    non_vectorized_times.append(non_vec_time)
plt.figure(figsize=(4, 2))
plt.plot(vectorized_times, marker='o', label=' ')
plt.plot(non_vectorized_times, marker='o', label=' ')
plt.title('       multisets')
plt.xlabel(' ')
plt.ylabel('  ()')
plt.legend()
plt.grid()
plt.show()

vectorized_times = []
non_vectorized_times = []

for data in max_prod_mod_3_data:
    vec_time = count_time(research_functions_vectorised.max_prod_mod_3, data)
    vectorized_times.append(vec_time)
    non_vec_time = count_time(research_functions.max_prod_mod_3, data.tolist())
    non_vectorized_times.append(non_vec_time)

plt.figure(figsize=(4, 2))
plt.plot([10, 100, 1000, 10000], vectorized_times, marker='o', label=' ')
plt.plot([10, 100, 1000, 10000], non_vectorized_times, marker='o', label=' ')
plt.title('       max_prod_mod_3')
plt.xlabel(' ')
plt.ylabel('  ()')
plt.legend()
plt.grid()
plt.show()


vectorized_times = []
non_vectorized_times = []

for data in convert_image_data:
    non_vec_time = count_time(research_functions.convert_image, data[0], data[1])
    non_vectorized_times.append(non_vec_time)
    vec_time = count_time(research_functions_vectorised.convert_image, data[0].tolist(), data[1].tolist())
    vectorized_times.append(vec_time)
plt.figure(figsize=(4, 2))
plt.plot(vectorized_times, marker='o', label=' ')
plt.plot(non_vectorized_times, marker='o', label=' ')
plt.title('       convert_image')
plt.xlabel(' ')
plt.ylabel('  ()')
plt.legend()
plt.grid()
plt.show()

vectorized_times = []
non_vectorized_times = []

for data in rle_scalar_data:
    vec_time = count_time(research_functions_vectorised.rle_scalar, data[0], data[1])
    vectorized_times.append(vec_time)
    non_vec_time = count_time(research_functions.rle_scalar, data[0].tolist(), data[1].tolist())
    non_vectorized_times.append(non_vec_time)
plt.figure(figsize=(4, 2))
plt.plot([10, 100, 1000, 10000], vectorized_times, marker='o', label=' ')
plt.plot([10, 100, 1000, 10000], non_vectorized_times, marker='o', label=' ')
plt.title('       rle_scalar')
plt.xlabel(' ')
plt.ylabel('  ()')
plt.legend()
plt.grid()
plt.show()

vectorized_times = []
non_vectorized_times = []

for data in cosine_distance_data:
    vec_time = count_time(research_functions_vectorised.cosine_distance, data[0], data[1])
    vectorized_times.append(vec_time)

    non_vec_time = count_time(research_functions.cosine_distance, data[0].tolist(), data[1].tolist())
    non_vectorized_times.append(non_vec_time)

plt.figure(figsize=(4, 2))
plt.plot([10, 100], vectorized_times, marker='o', label=' ')
plt.plot([10, 100], non_vectorized_times, marker='o', label=' ')
plt.title('       cosine_distance')
plt.xlabel(' ')
plt.ylabel('  ()')
plt.legend()
plt.grid()


````
